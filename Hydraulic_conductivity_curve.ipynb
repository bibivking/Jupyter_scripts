```python
import os
import sys
import math
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.colors
from matplotlib import ticker
import datetime as dt
```

```python
watr = 0.01
ssat = 0.4173705
hyds = 0.005727897*3600*24 # mm/day
#bch  = 5

K    = np.zeros((len(np.arange(watr + 0.01, ssat ,0.01)),13))

j = 0
for bch in np.arange(1.,14.,1):
    i = 0
    for theta in np.arange(watr + 0.01, ssat ,0.01):    
        theta_srf = (theta - watr)/(ssat - watr)
        K[i,j] = theta_srf**(2*bch+3) # m/day # 0.001*hyds*
        i += 1
    j += 1
    
fig = plt.figure(figsize=[15,10])
ax1 = fig.add_subplot(111)
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,0], color = 'black', label="b=1")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,1], color = 'gray', label="b=2")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,2], color = 'red', label="b=3")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,3], color = 'pink', label="b=4")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,4], color = 'orange', label="b=5")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,5], color = 'yellow', label="b=6")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,6], color = 'purple', label="b=7")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,7], color = 'blue', label="b=8")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,8], color = 'cyan', label="b=9")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,9], color = 'lightblue', label="b=10")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,10], color = 'green', label="b=11")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,11], color = 'olive', label="b=12")
ax1.plot(np.arange(watr + 0.01, ssat ,0.01), K[:,12], color = 'lightgreen', label="b=13")
ax1.set_yscale('log')
ax1.set_ylim(1e-15,1.)
ax1.legend()
```
